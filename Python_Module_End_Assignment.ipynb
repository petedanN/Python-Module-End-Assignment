# **Analysis Tasks:**

**1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees**

```python
#Determine the distribution of employees across each team.
team_count = df["Team"].value_counts()
team_count
```

```python
#Calculate the percentage split relative to the total number of employees
team_percentage = (team_count / df.shape[0]) * 100
team_df = pd.DataFrame({
    'Employee Count' : team_count,
    'Percentage' : team_percentage
})
team_df
```

```python
#Plot the employees count by team
plt.figure(figsize = (10, 6))
sns.barplot(x = team_count.index, y = team_df['Employee Count'])
plt.xticks(rotation = 90)
plt.title('Employee Distribution by Team')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.show()
```

From the Barplot we can see that the team 'New Orleans Pelicans' has highest number of employees (19).

**2. Segregate employees based on their positions within the company.**

```python
#Number of employees in each Position
position_distribution = df['Position'].value_counts()
print(position_distribution)
```

```python
#Plot employees count by position
plt.figure(figsize = (8, 6))
sns.barplot(x = position_distribution.index, y = position_distribution)
plt.title("Employees count by position")
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.show()
```

As we can see, Positon "SG" have the most number of employees (102).

**3. Identify the predominant age group among employees.**

```python
#Number of employees in each age
age_counts = df['Age'].value_counts()
print(age_counts)
```

```python
#Plot employees count by age
plt.figure(figsize = (8,6))
sns.countplot(x = 'Age', data = df)
plt.ylabel("Number of employees")
plt.title("Employees count by age")
plt.show()
```

Here we can see that, 47 employees have age 24, Which is the highest here.
Also we can see that, number of employees decreases as age increases.

**4. Discover which team and position have the highest salary expenditure.**

```python
salary_df = df[['Name','Team','Position','Salary']]
salary_df
```

```python
#Maximum salary
salary_sum = df.groupby(['Team', 'Position'])['Salary'].sum()
salary_sum
```

```python
max_salary_group = salary_sum.idxmax()  #team and position with highest salary
max_salary_group
```

```python
max_salary = salary_sum.max() #highest salary
max_salary
```

```python
salary_sum = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()
salary_sum
plt.figure(figsize = (10,8))
sns.scatterplot(data = salary_sum, x = 'Team', y = 'Salary', hue = 'Position')

plt.xlabel('Team')
plt.ylabel('Total Salary')
plt.xticks(rotation = 90)
plt.show()
```

From the Scatter Plot we can easily visualize Team vs Total Salary with Position.
The team Los Angeles Lakers's SF position is having the higest salary among others.

**5. Investigate if there's any correlation between age and salary, and represent it visually.**

```python
age_salary = df[['Age', 'Salary']]
age_salary
```

```python
corr = age_salary.corr()
print(corr)

sns.heatmap(corr, annot = True)
plt.title("Correlation between Age and Salary")
plt.show()
```

This correlation is to get a basic understanding of the Age and Salary relation.
As we can see here the relationship betweeen Age and Salary (0.21117) is positive relation but its actually a weak relationship.
